```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 

plt.style.use("default")
```

```python
df=pd.read_csv("C:/Users/agast/Downloads/apps.csv")
df.head()
```

## Dataset Overview 

The dataset contains information about Android applications available on the Google Play Store, including category, rating, number of installs, app size, and pricing details.

```python
df.shape
```

```python
df.info()
```

## Data Cleaning and Preparation

This step involves handling missing values, correcting data types, and cleaning inconsistent entries to ensure accurate analysis.

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

## Data Type Corrections 

Certain columns such as installs, price, annd size need to be cleaned and converted into numerical format.

```python
df['Installs'] = df['Installs'].str.replace(',', '').str.replace('+', '').astype(int)
df['Price'] = df['Price'].str.replace('$', '').astype(float)
```

## Category Exploration

```python
plt.figure(figsize=(10,5))
df['Category'].value_counts().plot(kind='bar')
plt.title("App Distribution Across Categories")
plt.xlabel("Category")
plt.ylabel("Number of Apps")
plt.show()
```

## Metrics Analysis

This section analyzes key metrics such as ratings, installs, and pricing trends across apps.

```python
sns.histplot(df['Rating'],bins=20)
plt.title("Distribution of App Ratings")
plt.show()
```

```python
plt.scatter(df['Rating'],df['Installs'])
plt.xlabel("Rating")
plt.ylabel("Installs")
plt.title("Installs vs Ratings")
plt.show()
```

## Pricing Analysis 

```python
sns.boxplot(x=df['Price'])
plt.title("App Price Disstribution")
plt.show()
```

## Key Insights

- Most apps on the Play Store are free
- Certain categories dominate the app market
- Higher-rated apps tend to have higher installs
- Paid apps are relatively fewer compared to free apps
